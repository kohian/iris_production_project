```python
import joblib
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
print(iris.DESCR)
```

```python
# create DataFrame with headers
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# add target column
df["target"] = iris.target

# add species names
df["species"] = df["target"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
```

```python
df.head()
```

```python
df.to_csv("iris.csv", index=False)
```
